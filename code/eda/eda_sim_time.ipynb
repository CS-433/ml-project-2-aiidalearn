```python
import os
import sys

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


ROOT_DIR = os.path.dirname(
os.path.dirname(os.path.abspath('')))


sys.path.append(os.path.join(ROOT_DIR, "code"))


# Define global variables
DATA_DIR = os.path.join(ROOT_DIR, "data/")

PLOT_DIR = os.path.join(ROOT_DIR, "plots/eda/")

DATA_PATH = os.path.join(DATA_DIR, "data.csv")
```

```python
data = pd.read_csv(DATA_PATH, index_col=0, na_filter=False)
```

```python
data.describe()
```

```python
blue = "#0063be"
yellow = "#fed900"
green = "#007836"
orange = "#ff6f20"
purple = "#580f8b"
```

```python
PLOT_PATH = os.path.join(PLOT_DIR, "sim_time.png")
os.makedirs(PLOT_DIR, exist_ok=True)
fs = 12
fig, ax = plt.subplots(1, 2, figsize=(6,3))
sns.kdeplot(data['time'], ax=ax[0], c=blue)
sns.kdeplot(np.log(data['time']), ax=ax[1], c=blue)

ax[0].set_xlabel("$Sim. time$", fontsize=fs)
ax[1].set_xlabel("$\\log(Sim. time$)", fontsize=fs)
plt.tight_layout()
plt.savefig(PLOT_PATH)
plt.show()
```
