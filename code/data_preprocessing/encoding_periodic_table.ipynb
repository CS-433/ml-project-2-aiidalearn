```python
import json
import os
import requests
```

```python
# LOADING ALL ELEMENT KEYS
url_table = requests.get("https://archive.materialscloud.org/record/file?record_id=862&filename=SSSP_1.1.2_PBE_efficiency.json&file_id=a5642f40-74af-4073-8dfd-706d2c7fccc2")
text_table = url_table.text
sssp_table = json.loads(text_table)
periodic_table_keys = list(sssp_table.keys())
```

```python
c1 = ['H', 'Li', 'Na', 'K', 'Rb', 'Cs']
c2 = ['Be', 'Mg', 'Ca', 'Sr', 'Ba']
c3 = ['Sc', 'Y']
c4 = ['Ti', 'Zr', 'Hf']
c5 = ['V', 'Nb', 'Ta']
c6 = ['Cr', 'Mo', 'W']
c7 = ['Mn', 'Tc', 'Re']
c8 = ['Fe', 'Ru', 'Os']
c9 = ['Co', 'Rh', 'Ir']
c10 = ['Ni', 'Pd', 'Pt']
c11 = ['Cu', 'Ag', 'Au']
c12 = ['Zn', 'Cd', 'Hg']
c13 = ['B', 'Al', 'Ga', 'In', 'Tl']
c14 = ['C', 'Si', 'Ge', 'Sn', 'Pb']
c15 = ['N', 'P', 'As', 'Sb', 'Bi']
c16 = ['O', 'S', 'Se', 'Te', 'Po']
c17 = ['F', 'Cl', 'Br', 'I']
c18 = ['He', 'Ne', 'Ar', 'Kr', 'Xe', 'Rn']
cols = [c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14, c15, c16, c17, c18]
lanthanides = ['La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu']
```

```python
periodic_table_info = {}
for element in periodic_table_keys:
    for col in cols:
        if element in col:
            periodic_table_info[element] = 'PTC' + str(cols.index(col) + 1)
    
for element in lanthanides:
    periodic_table_info[element] = 'Lant'
```

```python
colnames = ['PTC' + str(n) for n in range(1, 19)]
colnames.append('Lant')

for col in colnames:
    print("----", col, '----')
    for elt, ptc in periodic_table_info.items():
        if ptc == col:
            print(elt)
    
```

```python
filename = 'periodic_table_info'
```

```python
JSON_PATH = os.path.join(os.path.dirname(os.path.dirname(os.getcwd())), "data/" + filename + ".json")
with open(JSON_PATH, 'w') as write_file:
    json.dump(periodic_table_info, write_file)
```
